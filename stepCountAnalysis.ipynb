```python
import pandas as pd
import matplotlib.pyplot as plt

#make the plot size bigger
plt.rc('figure', figsize=(20, 10))  
step = pd.read_csv("stepCount.csv")
df=pd.DataFrame(step)
```

```python
df1 = df.loc[:, "date"]
df2 = df.loc[:, "step"]
date = []
step = []

#change date value from float to string in order to split
df1new = df1.apply(str)

#keep the date and remove the time
for x in df1new:
    x = x.split(" ")[0]
    date.append(x)

for x in df2:
    step.append(x)

df3 = pd.DataFrame({ "date" : date, 
                     "step": step       
})

#new dataframe that has the sum of step for every date
df4 = df3.groupby('date').sum()

#plot the graph of step report
df4.plot(color='red',linewidth=2.0)
plt.legend(prop={'size':25})
plt.xlabel("date",fontsize=25)
plt.ylabel("step",fontsize=25)
plt.title('step report 2020',fontsize=30)
plt.show()




```
